```python
#handling non-standard missing value
missing_values = ['s', 'na', 'n/a', 'N/A', '--']
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values = missing_values)

#Showing the first 10 rows of data
df.head(10)
```

```python
#Replacing missing values with 0
df['Number Scoring 65 or Above'].fillna(0,inplace=True)
df['Number Scoring Below 65'].fillna(0,inplace=True)
df['Mean Score'].fillna(0,inplace=True)
df['Number Scoring 80 or Above'].fillna(0,inplace=True)
df['Percent Scoring Below 65'].fillna(0,inplace=True)
df['Percent Scoring 65 or Above'].fillna(0,inplace=True)
df['Percent Scoring 80 or Above'].fillna(0,inplace=True)
df.head(10)
```

```python
##plot for the comparision between School#1 and School#2

#X-axis values
x = np.array(['2015', '2016', '2017'])

#Y-axis values
Liberty = [359, 529, 464] #District 2
Orchard = [188, 119, 244] #District 1

#Function to plot
x_axis = np.arange (len (x))
plt.bar(x_axis - 0.2, Liberty, 0.2, \
        label = 'Liberty High School Academy for Newcomers', color = 'purple')
plt.bar(x_axis + 0.2, Orchard, 0.2, \
        label = 'Orchard Collegiate Academy', color = 'green')

plt.xticks(x_axis, x) #set text labels
plt.xlabel('Years')
plt.ylabel('Total Tested')
plt.title('Number Scoring 80 or Above')
plt.legend(loc='center')
plt.show()
```

Analysis/Findings:

Analysis shows that Liberty High School had more students scoring 80 or above in all 3yrs, compared to Orchard Collegiate.
Data also shows that Liberty High School saw a dramatic increase in the number of students scoring 80 or above in 2016 compared to Orchard Collegiate, which saw a decrease in numbers compared to their 2015 numbers. Liberty High School saw their highest number of students scoring 80 or above in 2016. Orchard Collegiate saw their highest numbers in 2017, which was still lower that the lowest numbers had by Liberty High School, overall.

There was a decrease in numbers scoring 80 or above for Liberty High School in 2017, but the numbers were still higher than their 2015 numbers; and higher than every Orchard Collegiate numbers in every year provided.

The data provided for the number of all test participants shows an overall decline in the number of test participants in 2017, showing the lowest numbers of Test participants in all 3yrs provided.

When compared to the 2 schools, the overall data trend matches with Liberty High School except for 2017. Liberty High School performed better in 2017 than 2015, which was their lowest year; even with the 2017 having the lowest number of total test paricipants overall.

Orchard Collegiate data revealed the opposite, showing that there was a big decline in their numbers of students scoring 80 or above in 2016. Although, there were more test paricipants overall in 2016.
